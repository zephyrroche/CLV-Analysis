```python
data = pd.read_csv('customer_acquisition.csv')
data
```

# **Make a Visualization for the distribution of Customer Acquisition Cost.**

```python
fig = px.histogram(data,
                   x='cost',
                   nbins = 20,
                   title = 'Distribution of Acquisition Cost')
fig.show()
```

it shows a right-skewed distribution, with most costs falling under $20, though there are some higher outliers.

# **Create a Visualization for the revenue generated by the customer.**

```python
fig = px.histogram(data,
                   x='revenue',
                   nbins = 20,
                   title = 'Distribution of Revenue')
fig.show()
```

The revenue distribution is also right-skewed, with most customer revenues falling below $5,000 but some generating much higher revenue.

```python
data.head()
```

# **Compare the cost of acquisition across various channels and determine which ones are the most and least profitable?**

```python
cost_by_channel = data.groupby('channel')['cost'].mean().reset_index()

fig = px.bar(cost_by_channel, x='channel',
             y='cost',
             title='Customer Acquisition Cost by Channel')

fig.show()
```

email marketing is the least profitable way to gain customer acquisition while paid advertising is the most profitable.

# **Find out which channels are most and least effective at converting customers.**

```python
conversion_by_channel = data.groupby('channel')['conversion_rate'].mean().reset_index()

fig = px.bar(conversion_by_channel, x='channel',
             y='conversion_rate',
             title='Conversion Rate by Channel')

fig.show()
```

when it comes to converting customers social media is much more effective than paid advertising.

# **Calculate the total revenue by channel and analyze the most and the least profit channels in terms of generating revenue.**

```python
revenue_by_channel = data.groupby('channel')['revenue'].sum().reset_index()

fig = px.pie(revenue_by_channel,
             values='revenue',
             names='channel',
             title='Total Revenue by Channel',
             hole=0.6, color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show()
```

email marketing is the most profitable, social media is the least when it comes to generating revenue.

# **Calculate the return on investment (ROI) for each channel.**

```python
data['roi'] = data['revenue']/data['cost']
roi_by_channel = data.groupby('channel')['roi'].mean().reset_index()

fig = px.bar(roi_by_channel,
             x='channel',
             y='roi',
             title='Return on Investment (ROI) by Channel')

fig.show()
```

email marketing has an roi of 538.6175
